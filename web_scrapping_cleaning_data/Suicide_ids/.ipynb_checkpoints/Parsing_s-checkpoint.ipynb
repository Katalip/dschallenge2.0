```python
iterator = 0
is_open = True
wait = WebDriverWait(driver, 10)
for x in links:
    #Opening account
    driver.get(x)
    last_id = x
    
    try:
        check_open = driver.find_element_by_class_name('profile_closed_wall_dummy_title').text     
        is_open = False
    except NoSuchElementException:  #spelling error making this code not work as expected
        is_open = True
        
    
    if(is_open):
        #Initialize variables
        name = None; bday = None; city = None; university = None; status = None;
        friends = []; n_photos = []; groups_l = []; last_10_posts = []

        try:
            name = driver.find_element_by_class_name('page_name').text
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            bmonth = driver.find_element_by_xpath("//a[contains(@href, '[bday]')]").text
            byear = driver.find_element_by_xpath("//a[contains(@href, '[byear]')]").text
            bday = bmonth + ' ' + byear
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            city = driver.find_element_by_xpath("//a[contains(@href, '[city]')]").text
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            university = driver.find_element_by_xpath("//a[contains(@href, '[university]')]").text
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            n = driver.find_elements_by_xpath("//a[contains(@href, 'friends')]")
            for e in n:
                friends.append(e.text)
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            photos = driver.find_elements_by_xpath("//a[contains(@href, 'album')]")
            for x in photos:
                n_photos.append(x.text)
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            status = driver.find_element_by_class_name('current_text').text
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            element = driver.find_element_by_xpath("//*[contains(text(), 'Подписки')]")
            element.click()
            time.sleep(2)
            
            groups = driver.find_elements_by_class_name('fans_idol_lnk')
            for i in groups:
                groups_l.append(i.text) 
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass
        
        try:
            groups = driver.find_elements_by_class_name('fans_idol_lnk')
            for i in groups:
                groups_l.append(i.text) 
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            posts_l = driver.find_elements_by_class_name('wall_post_text')
            dates = driver.find_elements_by_class_name('rel_date')
            #dates = driver.find_elements_by_class_name('_post_content')
            j = 0
            for i in posts_l:
                last_10_posts.append({'text':i.get_attribute('innerText'), 'date':dates[j].get_attribute('innerText')})
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        df_final.loc[len(df_final)] = [name, bday, city, university, friends, n_photos, status, groups_l, last_10_posts]

    time.sleep(np.random.randint(10,30))
    iterator += 1
    
    if(iterator % 100 == 0):
        df_final.to_excel('vk_s'+str(iterator)+'_'+str(add_to_saving_name).split(' ')[0]+'.xlsx')
    print('{} accounts have been processed'.format(iterator), len(groups_l))
    
     
df_final.to_excel('vk_s'+str(iterator)+'_'+str(add_to_saving_name).split(' ')[0]+'.xlsx')
print('Success')
```
