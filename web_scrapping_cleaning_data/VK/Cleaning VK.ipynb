```python
df = pd.read_excel('data/raw_vk/vk_10_935_2019-11-177.xlsx', index_col = 0)
```

```python
def drop_row_all_nulls(df):
    df.drop(df[df['Name'].isnull()].index, inplace = True)
    df = df[df['Last_10_posts'] != '[]']
    return df
```

```python
def get_age(df):
    '''Extract age from birthday and fill NaN with mean value'''
    df['age'] = df[df['Bday'].notnull()]['Bday'].apply(lambda x: 2019 - int(x.split(' ')[2]))
    df['age'] = df['age'].fillna(np.round(df['age'].mean()))
    df['Bday'] = df['age']
    df.drop('age', axis=1, inplace=True)
    df.rename(columns = {'Bday':'Age'}, inplace=True)
    return df
```

```python
def f_city(df):
    df['City'] = df['City'].astype('object').fillna('Алматы')
    return df
```

```python
def university(df):
    df['University'] = df['University'].fillna(df.groupby('University')['Name'].count().sort_values(ascending=False).index[1])
    return df
```

```python
def N_friends(df):
    df['N_Friends'] = df['N_Friends'].apply(lambda x: x.split(' ')[-1].split('\\')[0].split("'")[-1])
    return df
```

```python
import re
def get_n_photos(df):
    df['N_Photos'] = df['N_Photos'].apply(lambda x: re.findall(r'\d+', x[x.find('фотографии')-7:x.find('фотографии')]))
    df['N_Photos'] = df['N_Photos'].apply(lambda x: 0 if not x else x[0])
    df['N_Photos'] = df['N_Photos'].astype('int') 
    return df
```

```python
def status(df):
    df['Status'] = df['Status'].fillna('Unfilled')
    return df
```

```python
df = drop_row_all_nulls(df)
```

```python
df = get_age(df)
```

```python
df = f_city(df)
```

```python
df = university(df)
```

```python
df = N_friends(df)
```

```python
df = get_n_photos(df)
```

```python
df = status(df)
```
