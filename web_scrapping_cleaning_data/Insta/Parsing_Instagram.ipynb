```python
iterator = 0
is_open = True
wait = WebDriverWait(driver, 10)
for x in insta_ids:
    #Opening account
    driver.get(x)
    last_id = x
    
    time.sleep(np.random.randint(5,10))
    
    if(is_open):
        #Initialize variables
        account_name = None; name = None; stats = None; bio = None;
        total_s = []; posts_l = []; posts_info_l = []; f_l = []

        try:
            account_name = driver.find_element_by_class_name('p-h2').text
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            name = driver.find_element_by_class_name('p-h1').text
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            stats = driver.find_element_by_class_name('p-stats-left').text
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            total_stats = driver.find_elements_by_class_name('profile-metric-item')
            for k in total_stats:
                total_s.append(k.get_attribute('innerText'))
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            bio = driver.find_element_by_class_name('p-bio').get_attribute('innerText')
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            posts = driver.find_elements_by_class_name('grid-media-caption')
            for k in posts:
                posts_l.append(k.get_attribute('innerText'))
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            posts_info = driver.find_elements_by_class_name('grid-media-info')
            for k in posts_info:
                posts_info_l.append(k.get_attribute('innerText'))
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass

        try:
            element = driver.find_element_by_xpath("//a[contains(@href, 'followings')]").get_attribute('href')
            driver.get(element)
            time.sleep(2)
            
            followings = driver.find_elements_by_class_name('grid-user-right')
        
            for k in followings:
                f_l.append(k.get_attribute('innerText')) 
        except NoSuchElementException:  #spelling error making this code not work as expected
            pass
        

        df_final.loc[len(df_final)] = [account_name, name, stats, bio, total_s, posts_l, posts_info_l, f_l]

    time.sleep(np.random.randint(5,20))
    iterator += 1
    
    if(iterator % 100 == 0):
        df_final.to_excel('data/raw_insta/insta_'+str(iterator)+'_'+str(add_to_saving_name).split(' ')[0]+'.xlsx')
    print('{} accounts have been processed'.format(iterator), len(f_l))
    
df_final.to_excel('data/raw_insta/insta_'+str(iterator)+'_'+str(add_to_saving_name).split(' ')[0]+'.xlsx')
print('Success')
```
